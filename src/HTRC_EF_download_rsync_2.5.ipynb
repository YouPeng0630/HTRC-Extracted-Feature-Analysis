#Generate Download Paths

```python
import csv
from htrc_features import utils
def normalize_htid(htid):
    # Fix CSV format: "ark_ 13960 xxx" -> "ark:/13960/xxx"
    if 'ark_' in htid:
        htid = htid.replace('ark_ ', 'ark:/')
        htid = htid.replace(' ', '/')
    else:
        htid = htid.replace(' ', '')
    return htid
# Generate paths
paths = []
with open('list.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        htid = normalize_htid(row['htid'])
        path = utils.id_to_rsync(htid, format='stubbytree')
        paths.append(path)
# Save to file
with open('file_paths.txt', 'w') as f:
    for path in paths:
        f.write(path + '\n')
print(f"✓ Generated {len(paths)} paths")
```
